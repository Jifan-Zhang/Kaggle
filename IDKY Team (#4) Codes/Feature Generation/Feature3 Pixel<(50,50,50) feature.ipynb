```
from google.colab import drive
drive.mount("/content/drive/")
from tqdm import tqdm
from PIL import Image
import numpy as np
import pickle

PATH="/content/drive/Shared drives/MyDrive/Kaggle/clips(enhanced) test/"
```

```
with open("/content/drive/Shared drives/MyDrive/Kaggle/clips(enhanced) test/_temp",'w') as f:
    f.write("25000")
```

```
with open("/content/drive/Shared drives/MyDrive/Kaggle/clips(enhanced) test/_temp",'r') as f:
    start=int(f.read())+1
print(f"Starting with {start} picture.")
data=pickle.load(open("/content/drive/Shared drives/MyDrive/Kaggle/Feature Generation (Enhanced)/Real size+grey+50 feature.pkl",'rb'))

for i in range(start,50001):
    filepath=PATH+"clips-"+str(i)+".png"
    count=np.apply_along_axis(lambda x: True if x.max()<50 else False, 2, np.array(Image.open(filepath)))
    data.loc[i,"50"]=count.sum()
    if(i%50==0):
        with open("/content/drive/Shared drives/MyDrive/Kaggle/clips(enhanced) test/_temp",'w') as f:
            f.write(str(i))
        pickle.dump(data,open("/content/drive/Shared drives/MyDrive/Kaggle/Feature Generation (Enhanced)/Real size+grey+50 feature.pkl",'wb'))
    if(i%500==0):
        print(f"{i} completed.")
```
